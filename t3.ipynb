```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({"prod_name":['g1','g2','c1','c2','c3','f1','f2'],
                "price":[45,25,75,35,89,20,55],
                "monthly_demand":[1200,2500,200,1850,120,3500,545],
                "brand":['z','x','x','y','z','y','z']})
```

```python
df.info()
```

```python
df
```

```python
df["monthly revenue"]=df['price']*df['monthly_demand']
df
```

```python
np.sum(df.loc[df['brand']=='z','monthly revenue'])
```

```python
df['revenue_share']=df['monthly revenue']
df
```

```python
df.loc[df['revenue_share']>=np.max(df['revenue_share']),'brand']
```
